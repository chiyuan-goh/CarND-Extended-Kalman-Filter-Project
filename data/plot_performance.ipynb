```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset1_output.csv")
```

```python
df
```

```python
both_arr = df.both.values
lidar_arr = df.lidar.values
radar_arr = df.radar.values
```

```python
both_arr.shape
```

```python
both_metrics = both_arr.reshape( (-1, 4))
lidar_metrics = lidar_arr.reshape( (-1, 4))
radar_metrics = radar_arr.reshape( (-1, 4))
```

```python
df_px = pd.DataFrame({"both": both_metrics[:, 0], "radar": radar_metrics[:, 0], "lidar":lidar_metrics[:, 0]})
df_py = pd.DataFrame({"both": both_metrics[:, 1], "radar": radar_metrics[:, 1], "lidar":lidar_metrics[:, 1]})
df_vx = pd.DataFrame({"both": both_metrics[:, 2], "radar": radar_metrics[:, 2], "lidar":lidar_metrics[:, 2]})
df_vy = pd.DataFrame({"both": both_metrics[:, 3], "radar": radar_metrics[:, 3], "lidar":lidar_metrics[:, 3]})
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16,14)
```

```python

df_px.plot()
plt.ylim(0,.01)
plt.title("plot of rms error of different sensors on x")
plt.savefig("../charts/x.png")
plt.show()
```

```python
df_py.plot()
plt.ylim(0,.01)
plt.title("plot of rms error of different sensors on y.")
plt.savefig("../charts/y.png")
plt.show()
```

```python
df_vx.plot()
plt.ylim(0,.05)
plt.title("plot of rms error of different sensors on vx")
plt.savefig("../charts/vx.png")
plt.show()
```

```python
df_vy.plot()
plt.ylim(0,.05)
plt.title("plot of rms error of different sensors on vy.")
plt.savefig("../charts/vy.png")
plt.show()
```
